```python
# PYTHON-EXPERIMENT-1
class Solution:

    def findMinDiff(self, arr, M):

        n = len(arr)

        # Edge cases
        if M == 0 or n == 0:
            return 0

        if M > n:
            return -1

        # Sort the array
        arr.sort()

        min_diff = float('inf')

        # Find the minimum difference
        for i in range(n - M + 1):
            diff = arr[i + M - 1] - arr[i]
            min_diff = min(min_diff, diff)

        return min_diff
```
